```python
import prepare
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import stats
```

```python
def plot_box(df):
    df = df.sample(1000)
    for col in list(df.columns):
        plt.title(col)
        sns.boxplot(data = df, y=col)
        plt.show()
        
def plot_variable_pairs(df, target):
    df = df.sample(1000)
    for col in list(df.columns):
        plt.title(col)
        sns.lmplot(data = df, x=target, y=col)
        plt.show()
        
def plot_categorical_and_continuous_vars(df,cat,cont):
    df = df.sample(1000)
    for col in cat:
        for c in cont:
            plt.title(col+ c)
            sns.swarmplot(x=col, y=c, data=df)
            plt.show()
            sns.stripplot(x=col, y=c, data=df)
            plt.show()
            sns.violinplot(x=col, y=c, data=df)
            plt.show()
```

Write a function named plot_categorical_and_continuous_vars that accepts your dataframe and the name of the columns that hold the continuous and categorical features and outputs 3 different plots for visualizing a categorical variable and a continuous variable.

```python
def plot_categorical_and_continuous_vars(df,cat,cont):
    df = df.sample(1000)
    for col in cat:
        for c in cont:
            plt.title(col+ c)
            sns.swarmplot(x=col, y=c, data=df)
            plt.show()
            sns.stripplot(x=col, y=c, data=df)
            plt.show()
            sns.violinplot(x=col, y=c, data=df)
            plt.show()
```

```python
cat = ['bedroomcnt','bathroomcnt','fips']
cont = ['calculatedfinishedsquarefeet','taxvaluedollarcnt','taxamount','yearbuilt']
```

Come up with some initial hypotheses based on your goal of predicting property value.

### 1. If you know the taxes and the area code you should know the property value

```python
sns.swarmplot(x='fips', y='taxvaluedollarcnt', data=train.sample(1000,random_state=21))
```

Are properties with fips of 6111 worth less than the others?

```python
overall_mean = train.taxvaluedollarcnt.mean()
alpha = .05
t, p = stats.ttest_1samp(train[train['fips']==6111.0]['taxvaluedollarcnt'], overall_mean)
print(t, p/2, alpha)
```

Fips of 6111 are worth less than the group mean

### 2. Bed+bath counts increase value
